## 二、各模块详细内容

### 模块一：量化交易框架与市场数据工程（3小时）

> 量化研究的第一步不是建模，是数据。数据质量决定了所有后续工作的上限。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第01课 | 量化交易的工作框架 | 量化全景图、买方vs卖方、策略分类、Alpha与Beta、Sharpe Ratio（夏普比率）、双均线策略实战 |
| 第02课 | 市场数据处理与清洗 | 幸存者偏差、前视偏差、复权处理、异常值检测（Z-score/MAD/Winsorize）、数据Pipeline |
| 第03课 | 特征工程与因子构建 | 技术因子（动量/波动率/换手率）、截面标准化、行业中性化、IC（信息系数）分析、五分位分组回测 |

### 模块二：因子研究与 Alpha 挖掘（3.5小时）

> 因子研究是量化投资的核心竞争力。好的因子研究不是发现一个"有效因子"，而是建立一套能持续甄别和迭代因子的系统。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第04课 | 单因子检验方法论 | IC/RankIC分析、因子衰减、Fama-MacBeth回归、Newey-West调整、多重检验（Bonferroni/FDR） |
| 第05课 | 多因子模型与Barra风险模型 | Barra因子体系、截面回归、因子协方差矩阵、风险分解、多因子合成（等权/IC加权/优化） |
| 第06课 | Fama-French扩展模型 | FF3/FF5模型、MKT/SMB/HML/RMW/CMA因子、Alpha提取、GRS检验、因子拥挤 |

### 模块三：时间序列建模与预测（3小时）

> 金融时间序列的核心挑战：信噪比极低，非平稳，尾部肥厚，波动聚类。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第07课 | 时间序列基础与平稳性检验 | Stylized Facts、ADF检验、ACF/PACF、ARIMA模型、AIC/BIC模型选择 |
| 第08课 | GARCH族波动率模型 | GARCH(1,1)、EGARCH、GJR-GARCH、波动率目标策略（Vol-Targeting）、VaR计算 |
| 第09课 | HMM与市场状态切换 | 隐马尔可夫模型（HMM）、Regime Switching、状态识别、动态策略切换 |

### 模块四：统计套利与配对交易（3小时）

> 统计套利的本质是找到偏离均衡的关系，并押注其回归。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第10课 | 协整理论与配对交易基础 | 协整vs相关性、Engle-Granger检验、Johansen检验、Z-score交易规则、半衰期 |
| 第11课 | Kalman Filter动态对冲 | 卡尔曼滤波器原理、动态对冲比率、预测-更新循环、与滚动OLS对比 |
| 第12课 | 误差修正模型与统计套利 | ECM模型、VECM、PCA统计套利、主成分分析、残差均值回复策略 |

### 模块五：机器学习在量化中的应用（3.5小时）

> ML不是魔法。在量化中，ML的价值在于处理非线性和高维交互。但过拟合是最大的敌人。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第13课 | ML建模框架与过拟合控制 | Walk-forward验证、Purged K-fold CV、过拟合演示、Deflated Sharpe Ratio |
| 第14课 | 线性模型、树模型与集成方法 | Ridge/Lasso/ElasticNet、Random Forest、LightGBM、特征重要性分析 |
| 第15课 | 贝叶斯方法在量化中的应用 | 先验/后验、贝叶斯收缩、贝叶斯线性回归、MCMC采样 |

### 模块六：波动率建模与衍生品量化（3小时）

> 期权市场交易的不是方向，是波动率。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第16课 | 期权定价与波动率曲面 | BSM模型、Greeks（Delta/Gamma/Vega/Theta）、隐含波动率、波动率曲面、Delta对冲 |
| 第17课 | 高级波动率模型与Copula | Heston随机波动率、Copula（Gaussian/t/Clayton/Gumbel）、尾部相关性 |
| 第18课 | Monte Carlo方法 | 蒙特卡洛定价、亚式/障碍期权、方差缩减技术、Monte Carlo VaR与压力测试 |

### 模块七：投资组合优化与风险管理（3小时）

> 好的组合管理不是选对股票，而是控制好风险暴露。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第19课 | 均值-方差优化与实际挑战 | MVO、有效前沿、Ledoit-Wolf收缩、正则化、约束优化 |
| 第20课 | Black-Litterman与风险平价 | BL模型、均衡收益、主观观点融合、风险平价、风险贡献 |
| 第21课 | 组合风险管理与归因 | VaR/CVaR三种方法、Brinson绩效归因、风控系统设计、压力测试 |

### 模块八：回测系统与交易执行（3小时）

> 回测是量化研究的"实验室"。一个好的回测系统能帮你发现错误，而不是帮你证明"策略有效"。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第22课 | 回测系统设计与常见偏差 | 前视偏差、幸存者偏差、交易成本建模、向量化回测框架 |
| 第23课 | 策略评估与样本外测试 | 评估指标体系、Bootstrap置信区间、参数敏感性、策略迭代 |
| 第24课 | 交易执行与实盘框架 | TWAP/VWAP/POV算法、订单簿分析、Almgren-Chriss模型、实盘框架设计 |

### 模块九：凸优化与高级建模专题（3小时）

> 把问题写成优化形式，是从"想法"到"可执行"的关键一步。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第25课 | 凸优化在量化中的应用 | LP/QP/SOCP、机构级组合优化器、交易成本感知再平衡、稳健优化 |
| 第26课 | PCA/SVD深入应用 | 随机矩阵理论（RMT）、协方差降噪、残差动量策略、Varimax旋转 |
| 第27课 | 强化学习基础框架 | RL基础（Agent/State/Action/Reward）、Q-Learning最优执行、与TWAP/VWAP对比 |

### 模块十：综合实战项目与课程总结（3小时）

> 把所有学到的东西串起来，像一个真正的量化研究员一样完成完整项目。

| 课程 | 主题 | 核心内容 |
|------|------|----------|
| 第28课 | 综合项目：多因子量化选股 | 端到端流程：数据→因子→检验→合成→优化→回测→归因→报告 |
| 第29课 | 综合项目：期权波动率策略 | GARCH预测→RV vs IV信号→Straddle交易→Delta对冲→P&L分解→执行模拟 |
| 第30课 | 课程总结与职业路径 | 知识图谱总结、量化职业路径、进阶方向、推荐阅读 |

## 六、常见问题（FAQ）

### Q1: `conda env create` 报错怎么办？

**尝试更新 conda 后重试：**
```bash
conda update -n base -c defaults conda
conda env create -f environment.yml
```

### Q2: Jupyter 中看不到 `Quant Finance` kernel？

**手动注册 kernel：**
```bash
conda activate quant-finance
python -m ipykernel install --user --name quant-finance --display-name "Quant Finance (Python 3.11)"
```

### Q3: 某个库 import 报错？

**在激活环境后手动安装：**
```bash
conda activate quant-finance
pip install <库名>
```

### Q4: 如何在另一台电脑上复现环境？

**只需要复制 `environment.yml` 文件，然后运行：**
```bash
conda env create -f environment.yml
conda activate quant-finance
jupyter notebook
```

### Q5: 如何删除环境重新安装？

```bash
conda deactivate
conda env remove -n quant-finance
conda env create -f environment.yml
```
