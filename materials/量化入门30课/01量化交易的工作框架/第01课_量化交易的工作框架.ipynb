## 1. 量化交易全景图 — 从研究到实盘的完整链路

量化交易是一个系统工程，涉及多个环节的协同工作。下面是一个典型的量化交易工作流：

```
┌──────────┐    ┌──────────┐    ┌──────────┐    ┌──────────┐    ┌──────────┐
│  数据获取 │───▶│  因子研究  │───▶│  回测验证 │─-─▶│  模拟盘  │──-> │   实盘    │
│  Data    │    │  Alpha   │    │ Backtest │    │  Paper   │    │  Live    │
└──────────┘    └──────────┘    └──────────┘    └──────────┘    └──────────┘
      │               │               │               │               │
      ▼               ▼               ▼               ▼               ▼
  市场数据         信号生成         绩效评估         实时监控         风险管理
  另类数据         因子组合         风险分析         滑点分析         头寸管理
  基本面数据       模型训练         归因分析         延迟测试         订单执行
```

### 各环节详解：

1. **数据获取与清洗**：获取行情、基本面、另类数据（新闻、卫星图等），清洗异常值，处理缺失数据
2. **因子研究与信号生成**：构建预测因子，挖掘Alpha信号
3. **回测验证**：在历史数据上模拟策略表现，注意避免过拟合和前视偏差
4. **模拟盘验证**：在实时市场中用虚拟资金运行策略，验证延迟、滑点等实盘问题
5. **实盘交易**：真实资金运行，持续监控风险和绩效

## 2. 买方与卖方量化团队的分工

### 买方（Buy-Side）

买方机构包括对冲基金、公募基金、养老基金等，目标是**获取投资收益**。

| 角色 | 职责 |
|------|------|
| 量化研究员（Quant Researcher） | 开发Alpha信号、构建因子、统计建模 |
| 量化开发（Quant Developer） | 搭建回测框架、数据管道、实盘系统 |
| 组合经理（Portfolio Manager） | 策略组合配置、风险预算、最终交易决策 |
| 风控（Risk Manager） | 监控组合风险、设置限额、压力测试 |

### 卖方（Sell-Side）

卖方机构包括投资银行、券商等，目标是**提供流动性和服务**。

| 角色 | 职责 |
|------|------|
| 量化分析师 | 衍生品定价、风险模型、策略研究报告 |
| 做市商（Market Maker） | 提供买卖报价，赚取价差 |
| 算法交易（Algo Trading） | 开发执行算法（TWAP, VWAP, IS等） |
| 电子交易（e-Trading） | 构建交易基础设施、智能路由 |

### 关键区别

- 买方关注 **预测**（forecast）—— 未来价格会如何变动？
- 卖方关注 **执行**（execution）—— 如何以最小成本完成交易？

## 3. 策略分类

### 3.1 统计套利（Statistical Arbitrage）

利用统计关系在多个资产之间进行配对或组合交易。

- **配对交易**：做多被低估的股票，做空被高估的股票（如可口可乐 vs 百事可乐）
- **截面动量/反转**：根据过去收益率的排名进行多空组合
- **特点**：市场中性、高换手率、依赖大量标的分散风险

### 3.2 趋势跟踪（Trend Following）

捕捉价格的持续性趋势，常用于期货市场（CTA策略，即商品交易顾问 Commodity Trading Advisor，是一类专门使用趋势跟踪策略的基金）。

- **均线突破**：价格突破均线时建仓
- **通道突破**：价格突破N日高点/低点时建仓（海龟交易法）
- **特点**：胜率低但盈亏比高、在趋势市场中表现好、与股票市场低相关

### 3.3 做市（Market Making）

同时挂出买卖双边报价，赚取买卖价差（Bid-Ask Spread）。

- **特点**：高频、低延迟、对基础设施要求极高
- **风险**：逆向选择风险（被知情交易者"pick off"）

### 3.4 事件驱动（Event-Driven）

基于特定事件（财报、并购、政策等）进行交易。

- **财报事件**：利用财报发布前后的价格异常
- **并购套利**：在宣布收购后买入标的公司
- **特点**：依赖对事件的快速解读、持仓时间较短

## 4. Alpha vs Beta，策略生命周期

### Alpha 与 Beta

- **Beta（β）**：市场系统性风险的暴露，可以通过指数基金低成本获取
- **Alpha（α）**：超越市场基准的超额收益，是量化研究的核心目标

$$R_p = \alpha + \beta \cdot R_m + \epsilon$$

其中 $R_p$ 是组合收益，$R_m$ 是市场收益，$\alpha$ 是超额收益，$\beta$ 是市场暴露。

### 策略生命周期

每个Alpha信号都有"保质期"，典型的生命周期如下：

```
收益
 ▲
 │     ╭───╮
 │    ╱     ╲         ← 衰减期：竞争加剧
 │   ╱       ╲
 │  ╱   黄金期  ╲──────── 长尾期：微薄收益
 │ ╱             ╲_______________
 │╱ 发现期
 ┼──────────────────────────────▶ 时间
```

- **发现期**：少数人发现Alpha信号，收益最高
- **黄金期**：策略成熟，风险调整后收益稳定
- **衰减期**：更多人发现同样的信号，Alpha被竞争"吃掉"
- **长尾期**：信号仍有微弱效果，但容量有限

> **量化团队的核心挑战**：持续寻找新的Alpha来源，同时管理现有Alpha的衰减。

## 5. 核心绩效指标

### 5.1 夏普比率（Sharpe Ratio，衡量每承受一单位风险获得多少收益）

> **通俗理解**：夏普比率就像"性价比"——同样是赚10%，波动小的策略夏普比率更高，说明赚钱更"稳"。

衡量单位风险下的超额收益：

$$\text{Sharpe Ratio} = \frac{E[R_p - R_f]}{\sigma_p}$$

- $R_p$：组合收益率
- $R_f$：无风险利率
- $\sigma_p$：组合收益率的标准差

**年化夏普比率** = 日夏普 × $\sqrt{252}$

| 夏普比率 | 评价 |
|---------|------|
| < 0.5 | 较差 |
| 0.5 - 1.0 | 一般 |
| 1.0 - 2.0 | 优秀 |
| > 2.0 | 卓越（需警惕过拟合） |

### 5.2 信息比率与基本定律

**信息比率（Information Ratio，IR，衡量相对于基准的风险调整后超额收益）**：

$$IR = \frac{E[R_p - R_b]}{\sigma_{R_p - R_b}}$$

其中 $R_b$ 是基准收益。

**Grinold主动管理基本定律**：

$$IR = IC \times \sqrt{\text{Breadth}}$$

- **IC（信息系数，Information Coefficient）**：预测能力，即因子预测值与实际收益的相关系数，衡量预测的准确度
- **Breadth（广度）**：独立预测的数量，即策略覆盖的独立投资机会数量（如选股数量 × 交易频率）

**启示**：
- 即使IC很小（如0.05），只要Breadth足够大，也能获得不错的IR
- IC=0.05, Breadth=500 → IR = 0.05 × √500 ≈ 1.12
- 这就是为什么量化基金倾向于交易大量标的、高频率调仓

````python
# 尝试使用yfinance获取真实数据，失败时使用合成数据
def get_stock_data(tickers, start='2020-01-01', end='2024-12-31'):
    """获取股票数据，支持yfinance和合成数据回退"""
    try:
        import yfinance as yf
        data = yf.download(tickers, start=start, end=end)['Close']
        if isinstance(data, pd.Series):
            data = data.to_frame(name=tickers if isinstance(tickers, str) else tickers[0])
        print(f"成功从yfinance获取数据，共 {len(data)} 个交易日")
        return data
    except Exception as e:
        print(f"yfinance不可用 ({e})，使用合成数据")
        return generate_synthetic_data(tickers, start, end)


def generate_synthetic_data(tickers, start='2020-01-01', end='2024-12-31'):
    """生成模拟股票价格数据
    
    使用GBM（几何布朗运动，Geometric Brownian Motion）模型：
    一种模拟股价随机变化的经典数学模型，假设股价的变化率服从正态分布"""
    np.random.seed(42)
    if isinstance(tickers, str):
        tickers = [tickers]
    
    dates = pd.bdate_range(start=start, end=end)  # 仅工作日
    n = len(dates)
    
    # 不同资产的参数设置
    params = {
        'SPY':  {'mu': 0.10, 'sigma': 0.18, 'S0': 320},
        'QQQ':  {'mu': 0.15, 'sigma': 0.25, 'S0': 210},
        'IWM':  {'mu': 0.07, 'sigma': 0.22, 'S0': 150},
        'AAPL': {'mu': 0.20, 'sigma': 0.30, 'S0': 75},
        'MSFT': {'mu': 0.18, 'sigma': 0.28, 'S0': 160},
    }
    
    data = {}
    for ticker in tickers:
        p = params.get(ticker, {'mu': 0.10, 'sigma': 0.20, 'S0': 100})
        dt = 1/252
        # GBM（几何布朗运动）公式: dS = mu*S*dt + sigma*S*dW
        # mu=年化收益率, sigma=年化波动率, dW=随机扰动
        log_returns = (p['mu'] - 0.5 * p['sigma']**2) * dt + \
                      p['sigma'] * np.sqrt(dt) * np.random.randn(n)
        prices = p['S0'] * np.exp(np.cumsum(log_returns))
        data[ticker] = prices
    
    df = pd.DataFrame(data, index=dates)
    print(f"已生成合成数据，共 {len(df)} 个交易日，标的: {tickers}")
    return df


# 获取数据
# SPY = SPDR S&P 500 ETF（追踪标普500指数，代表美国大盘股整体表现）
# QQQ = Invesco QQQ Trust（追踪纳斯达克100指数，以科技股为主）
# IWM = iShares Russell 2000 ETF（追踪罗素2000指数，代表美国小盘股）
# 这三个ETF（交易所交易基金）是美国市场最常用的宽基指数基金
tickers = ['SPY', 'QQQ', 'IWM']
prices = get_stock_data(tickers, start='2020-01-01', end='2024-12-31')
prices.head(10)
````

### 6.3 计算滚动夏普比率与最大回撤

````python
def rolling_sharpe(returns, window=252, rf_annual=0.02):
    """计算滚动夏普比率
    
    Parameters:
        returns: 日收益率序列
        window: 滚动窗口（交易日数）
        rf_annual: 年化无风险利率
    Returns:
        滚动年化夏普比率序列
    """
    rf_daily = rf_annual / 252
    excess_returns = returns - rf_daily
    rolling_mean = excess_returns.rolling(window).mean()
    rolling_std = excess_returns.rolling(window).std()
    return (rolling_mean / rolling_std) * np.sqrt(252)


def max_drawdown(prices_series):
    """计算最大回撤序列
    
    Parameters:
        prices_series: 价格序列
    Returns:
        回撤序列（负值表示回撤幅度）
    """
    cummax = prices_series.cummax()
    drawdown = (prices_series - cummax) / cummax
    return drawdown


def max_drawdown_value(prices_series):
    """计算最大回撤数值"""
    dd = max_drawdown(prices_series)
    return dd.min()


# 计算滚动夏普
sharpe_df = pd.DataFrame()
for col in daily_returns.columns:
    sharpe_df[col] = rolling_sharpe(daily_returns[col], window=252)

# 计算回撤
dd_df = pd.DataFrame()
for col in prices.columns:
    dd_df[col] = max_drawdown(prices[col])

# 可视化
fig, axes = plt.subplots(2, 1, figsize=(14, 8))

# 滚动夏普
ax = axes[0]
for col in sharpe_df.columns:
    ax.plot(sharpe_df[col], label=col, linewidth=1.2)
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
ax.axhline(y=1, color='green', linestyle='--', linewidth=0.8, alpha=0.5, label='Sharpe=1')
ax.axhline(y=-1, color='red', linestyle='--', linewidth=0.8, alpha=0.5, label='Sharpe=-1')
ax.set_title('252日滚动年化夏普比率', fontsize=13)
ax.legend(loc='upper right')
ax.grid(alpha=0.3)

# 回撤
ax = axes[1]
for col in dd_df.columns:
    ax.fill_between(dd_df.index, dd_df[col] * 100, 0, alpha=0.3, label=col)
ax.set_title('历史回撤（%）', fontsize=13)
ax.legend(loc='lower right')
ax.grid(alpha=0.3)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

plt.tight_layout()
plt.show()

# 打印最大回撤
print("\n最大回撤:")
for col in prices.columns:
    mdd = max_drawdown_value(prices[col])
    print(f"  {col}: {mdd:.2%}")
````

### 6.4 实现简单双均线策略

**关键概念：**
- **MA20** = 20日移动平均线（Moving Average），即过去20个交易日收盘价的平均值，代表**短期趋势**
- **MA60** = 60日移动平均线，即过去60个交易日（约3个月）的平均值，代表**中期趋势**
- 移动平均线就像"平滑后的价格走势"，去掉了日常波动的噪音，帮助我们看清价格的大方向

**策略逻辑：**
- 当短期均线（MA20）上穿长期均线（MA60）时 → **买入**（金叉，说明短期走势开始强于中期走势，趋势向上）
- 当短期均线（MA20）下穿长期均线（MA60）时 → **卖出**（死叉，说明短期走势弱于中期走势，趋势向下）
- 仅做多，不做空

这是最经典的趋势跟踪策略之一，虽然简单，但包含了策略构建的核心要素。

````python
def dual_ma_strategy(prices_series, short_window=20, long_window=60):
    """双均线策略
    
    Parameters:
        prices_series: 价格序列
        short_window: 短期均线窗口
        long_window: 长期均线窗口
    Returns:
        DataFrame包含价格、均线、信号和策略收益
    """
    df = pd.DataFrame(index=prices_series.index)
    df['price'] = prices_series
    df['returns'] = prices_series.pct_change()
    
    # 计算均线
    df['MA_short'] = prices_series.rolling(short_window).mean()
    df['MA_long'] = prices_series.rolling(long_window).mean()
    
    # 生成信号：短均线在长均线之上 → 持有（1），否则 → 空仓（0）
    df['signal'] = 0
    df.loc[df['MA_short'] > df['MA_long'], 'signal'] = 1
    
    # 注意：信号需要延迟一天执行（避免前视偏差）
    df['position'] = df['signal'].shift(1)
    
    # 策略收益 = 持仓 × 市场收益
    df['strategy_returns'] = df['position'] * df['returns']
    
    # 累积收益
    df['cum_market'] = (1 + df['returns']).cumprod()
    df['cum_strategy'] = (1 + df['strategy_returns']).cumprod()
    
    return df.dropna()


# 对SPY运行策略
spy_prices = prices['SPY']
result = dual_ma_strategy(spy_prices, short_window=20, long_window=60)

print(f"策略回测结果 (SPY, MA20/MA60):")
print(f"回测区间: {result.index[0].strftime('%Y-%m-%d')} 至 {result.index[-1].strftime('%Y-%m-%d')}")
print(f"总交易日数: {len(result)}")
print(f"持仓天数占比: {result['position'].mean():.1%}")
result.tail()
````

````python
# 可视化策略表现
fig, axes = plt.subplots(3, 1, figsize=(14, 12), 
                         gridspec_kw={'height_ratios': [3, 1, 2]})

# 1. 价格与均线
ax = axes[0]
ax.plot(result['price'], label='SPY', color='black', linewidth=1)
ax.plot(result['MA_short'], label='MA20（20日短期均线）', color='blue', linewidth=0.8, alpha=0.8)
ax.plot(result['MA_long'], label='MA60（60日中期均线）', color='red', linewidth=0.8, alpha=0.8)

# 标记买卖信号
buy_signals = result[result['signal'].diff() == 1]
sell_signals = result[result['signal'].diff() == -1]
ax.scatter(buy_signals.index, buy_signals['price'], marker='^', color='green', 
           s=80, zorder=5, label='买入信号')
ax.scatter(sell_signals.index, sell_signals['price'], marker='v', color='red', 
           s=80, zorder=5, label='卖出信号')

# 在第一个金叉点和死叉点添加文字标注，帮助理解策略逻辑
if len(buy_signals) > 0:
    first_buy = buy_signals.iloc[0]
    ax.annotate('\u2190 金叉：短期均线上穿长期均线，买入信号',
                xy=(buy_signals.index[0], first_buy['price']),
                xytext=(40, 30), textcoords='offset points',
                fontsize=9, color='green', fontweight='bold',
                arrowprops=dict(arrowstyle='->', color='green', lw=2))

if len(sell_signals) > 0:
    first_sell = sell_signals.iloc[0]
    ax.annotate('\u2190 死叉：短期均线下穿长期均线，卖出信号',
                xy=(sell_signals.index[0], first_sell['price']),
                xytext=(40, -40), textcoords='offset points',
                fontsize=9, color='red', fontweight='bold',
                arrowprops=dict(arrowstyle='->', color='red', lw=2))

ax.set_title('双均线策略 — SPY (MA20 / MA60)', fontsize=14)
ax.legend(loc='upper left')
ax.grid(alpha=0.3)

# 2. 持仓状态
ax = axes[1]
ax.fill_between(result.index, result['position'], 0, alpha=0.3, color='blue')
ax.set_ylabel('持仓')
ax.set_title('持仓状态（1=满仓, 0=空仓）', fontsize=12)
ax.set_ylim(-0.1, 1.1)
ax.grid(alpha=0.3)

# 3. 累积收益对比
ax = axes[2]
ax.plot(result['cum_market'], label='买入持有（Buy & Hold）', linewidth=1.5)
ax.plot(result['cum_strategy'], label='双均线策略', linewidth=1.5)
ax.set_title('累积收益对比', fontsize=13)
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
````

### 6.5 计算完整绩效指标

````python
def calc_performance_metrics(returns, rf_annual=0.02, periods_per_year=252):
    """计算全面的策略绩效指标
    
    Parameters:
        returns: 日收益率序列
        rf_annual: 年化无风险利率
        periods_per_year: 每年交易日数
    Returns:
        dict: 包含各项绩效指标
    """
    returns = returns.dropna()
    rf_daily = rf_annual / periods_per_year
    
    # 基本统计
    total_return = (1 + returns).prod() - 1
    n_years = len(returns) / periods_per_year
    annual_return = (1 + total_return) ** (1 / n_years) - 1
    annual_vol = returns.std() * np.sqrt(periods_per_year)
    
    # 夏普比率
    excess_returns = returns - rf_daily
    sharpe = excess_returns.mean() / excess_returns.std() * np.sqrt(periods_per_year) if excess_returns.std() > 0 else 0
    
    # 最大回撤
    cum = (1 + returns).cumprod()
    peak = cum.cummax()
    dd = (cum - peak) / peak
    mdd = dd.min()
    
    # 最大回撤持续时间
    underwater = dd < 0
    if underwater.any():
        groups = (~underwater).cumsum()
        dd_durations = underwater.groupby(groups).sum()
        max_dd_duration = dd_durations.max() if len(dd_durations) > 0 else 0
    else:
        max_dd_duration = 0
    
    # 卡尔马比率（Calmar Ratio）= 年化收益 ÷ 最大回撤，衡量收益与最大亏损的关系
    calmar = annual_return / abs(mdd) if mdd != 0 else np.inf
    
    # 索提诺比率（Sortino Ratio）：只惩罚下行波动的夏普比率改进版，不惩罚向上的波动
    downside_returns = returns[returns < 0]
    downside_vol = downside_returns.std() * np.sqrt(periods_per_year) if len(downside_returns) > 0 else 0
    sortino = (annual_return - rf_annual) / downside_vol if downside_vol > 0 else np.inf
    
    # 胜率
    win_rate = (returns > 0).mean()
    
    # 盈亏比
    avg_win = returns[returns > 0].mean() if (returns > 0).any() else 0
    avg_loss = abs(returns[returns < 0].mean()) if (returns < 0).any() else 1
    profit_loss_ratio = avg_win / avg_loss if avg_loss > 0 else np.inf
    
    return {
        '总收益率': f'{total_return:.2%}',
        '年化收益率': f'{annual_return:.2%}',
        '年化波动率': f'{annual_vol:.2%}',
        '夏普比率': f'{sharpe:.3f}',
        '索提诺比率(Sortino)': f'{sortino:.3f}',
        '最大回撤': f'{mdd:.2%}',
        '最大回撤持续(天)': f'{max_dd_duration:.0f}',
        '卡尔马比率(Calmar)': f'{calmar:.3f}',
        '日胜率': f'{win_rate:.2%}',
        '盈亏比': f'{profit_loss_ratio:.3f}',
        '交易日数': len(returns),
    }


# 计算对比指标
market_metrics = calc_performance_metrics(result['returns'])
strategy_metrics = calc_performance_metrics(result['strategy_returns'])

comparison = pd.DataFrame({
    '买入持有': market_metrics,
    '双均线策略': strategy_metrics
})

print("=" * 60)
print("策略绩效对比")
print("=" * 60)
print(comparison.to_string())
````

````python
# 信息比率（IR, Information Ratio）的直观演示
print("=" * 60)
print("Grinold基本定律: IR（信息比率） = IC（信息系数，预测准确度） × sqrt(Breadth（广度，独立投资机会数量））")
print("=" * 60)

# 模拟不同IC和Breadth组合
ic_values = [0.02, 0.05, 0.10, 0.15]
breadth_values = [50, 100, 250, 500, 1000, 2500]

ir_table = pd.DataFrame(index=[f'IC={ic}' for ic in ic_values],
                        columns=[f'N={b}' for b in breadth_values])

for i, ic in enumerate(ic_values):
    for j, breadth in enumerate(breadth_values):
        ir = ic * np.sqrt(breadth)
        ir_table.iloc[i, j] = f'{ir:.2f}'

print("\nIR = IC × √Breadth 对照表:")
print(ir_table)
print("\n说明: IC是预测能力（因子与收益的相关系数），Breadth是独立预测次数")
print("即使IC很小(0.05)，足够多的独立预测(N=500)也能产生优秀的IR(1.12)")

# 可视化
fig, ax = plt.subplots(figsize=(10, 6))
for ic in ic_values:
    breadths = np.arange(10, 3000)
    irs = ic * np.sqrt(breadths)
    ax.plot(breadths, irs, label=f'IC = {ic}', linewidth=2)

ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, label='IR=0.5 (一般)')
ax.axhline(y=1.0, color='green', linestyle='--', alpha=0.5, label='IR=1.0 (优秀)')
ax.axhline(y=2.0, color='gold', linestyle='--', alpha=0.5, label='IR=2.0 (卓越)')
ax.set_xlabel('Breadth（独立预测次数）', fontsize=12)
ax.set_ylabel('Information Ratio', fontsize=12)
ax.set_title('Grinold基本定律: IR = IC × √Breadth', fontsize=14)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
ax.set_xlim(0, 3000)
ax.set_ylim(0, 4)
plt.tight_layout()
plt.show()
````

### 6.6 参数敏感性分析

不同的均线参数组合会带来不同的策略表现。下面我们遍历多组参数来观察策略的稳健性。

````python
# 参数敏感性分析
short_windows = [5, 10, 20, 30, 40]
long_windows = [30, 60, 90, 120, 180]

# 存储结果
sharpe_matrix = np.zeros((len(short_windows), len(long_windows)))
return_matrix = np.zeros((len(short_windows), len(long_windows)))

for i, sw in enumerate(short_windows):
    for j, lw in enumerate(long_windows):
        if sw >= lw:
            sharpe_matrix[i, j] = np.nan
            return_matrix[i, j] = np.nan
            continue
        res = dual_ma_strategy(spy_prices, short_window=sw, long_window=lw)
        strat_ret = res['strategy_returns'].dropna()
        if len(strat_ret) > 0 and strat_ret.std() > 0:
            sharpe_matrix[i, j] = strat_ret.mean() / strat_ret.std() * np.sqrt(252)
            return_matrix[i, j] = ((1 + strat_ret).prod() ** (252/len(strat_ret)) - 1) * 100
        else:
            sharpe_matrix[i, j] = np.nan
            return_matrix[i, j] = np.nan

# 热力图
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 夏普比率热力图
ax = axes[0]
im = ax.imshow(sharpe_matrix, cmap='RdYlGn', aspect='auto')
ax.set_xticks(range(len(long_windows)))
ax.set_xticklabels(long_windows)
ax.set_yticks(range(len(short_windows)))
ax.set_yticklabels(short_windows)
ax.set_xlabel('长期均线窗口', fontsize=12)
ax.set_ylabel('短期均线窗口', fontsize=12)
ax.set_title('年化夏普比率', fontsize=13)
for i in range(len(short_windows)):
    for j in range(len(long_windows)):
        if not np.isnan(sharpe_matrix[i, j]):
            ax.text(j, i, f'{sharpe_matrix[i,j]:.2f}', ha='center', va='center', fontsize=9)
plt.colorbar(im, ax=ax)

# 年化收益热力图
ax = axes[1]
im = ax.imshow(return_matrix, cmap='RdYlGn', aspect='auto')
ax.set_xticks(range(len(long_windows)))
ax.set_xticklabels(long_windows)
ax.set_yticks(range(len(short_windows)))
ax.set_yticklabels(short_windows)
ax.set_xlabel('长期均线窗口', fontsize=12)
ax.set_ylabel('短期均线窗口', fontsize=12)
ax.set_title('年化收益率(%)', fontsize=13)
for i in range(len(short_windows)):
    for j in range(len(long_windows)):
        if not np.isnan(return_matrix[i, j]):
            ax.text(j, i, f'{return_matrix[i,j]:.1f}', ha='center', va='center', fontsize=9)
plt.colorbar(im, ax=ax)

plt.suptitle('双均线策略参数敏感性分析 (SPY)', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()

print("\n提示: 如果不同参数组合的表现差异很大，说明策略对参数敏感，需要谨慎。")
print("好的策略应该在较宽的参数范围内都有稳定的正收益。")
````
